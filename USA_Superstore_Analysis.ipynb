```python
# Filling missing values.
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Order Priority'].value_counts()
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority']=df['Order Priority'].replace('Critical ','Critical')
```

```python
df['Order Priority'].value_counts()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x='Order Priority',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Count of Order Priority')
plt.show()
# plt.savefig('Count of Order Priority.jpg')
```

```python
df['Ship Mode'].value_counts()
```

```python
x=df['Ship Mode'].value_counts().index
y=df['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,autopct='%0.2f%%',labels=x,startangle=60)
# plt.legend(loc=2)
plt.show()
```

```python
ax=sns.countplot(x='Order Priority',data=df,hue='Product Category')
for bars in ax.containers:
    ax.bar_label(bars)
```
